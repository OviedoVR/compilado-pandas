# Agregações

```python
# biblioteca:
import pandas as pd

# dados:
vendas = pd.read_csv('EuropeSalesRecords.csv')
vendas.head(3)
```

### **1. Método groupby()**

> Qual foi o faturamento total por país? (usemos 1 casas decimal)

```python
vendas.groupby('Country')[['Total Revenue']].sum().head()
```

```python
receitas_pais = vendas.groupby('Country', as_index=False)[['Total Revenue']].sum().sort_values('Total Revenue', ascending=False).round(1)
receitas_pais
```

> Qual foi o lucro médio em cada país por canal de venda? listemos o top 5

```python
lucro_medio_pais_canal = vendas.groupby(['Country', 'Sales Channel'])[['Total Profit']].mean()
lucro_medio_pais_canal = lucro_medio_pais_canal.sort_values('Total Profit', ascending=False).round(1)
lucro_medio_pais_canal.head()
```

### **2. Método agg**

```python
# lucro por país e canal de venda (média, std e mediana):
vendas.groupby(['Country', 'Sales Channel'])['Total Profit'].agg(['mean', 'std', 'median'])
```

```python
# agregação com `.describe()`:
vendas.groupby(['Sales Channel', 'Country'])[['Total Profit']].describe()
```
