### **6. Contagem de valores**

```python
# contagem para `Item Type`:
vendas['Item Type'].value_counts()
```

> Incluir nulos: opção `dropna=False`.

```python
# coluna `Order Priority`:
vendas['Order Priority'].value_counts(dropna=False)
```

####*Dica**
> Podemos inclusive plotar essa contagem, mas isso veremos com mais calma adiante. Vejamos um exemplo com `Item Type`

```python
vendas['Item Type'].value_counts(dropna=False).plot.barh(figsize=(5,4), color='black')
```
